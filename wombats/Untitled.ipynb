```python
import numpy as np
import seaborn as sns
```

```python
X = np.arange(100).reshape(10,10).T
sns.heatmap(X)
```

```python
n_neurons = 10
pct_neurons = 0.5

n_pert_neurons = int(n_neurons*pct_neurons)
pert_neurons = np.random.permutation(n_neurons)[:n_pert_neurons]

X2 = np.array(X)
X2[:,pert_neurons] = X2[:,np.roll(pert_neurons,1)]
```

```python
sns.heatmap(X2)
```
